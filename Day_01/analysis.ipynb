```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('production_log.csv') # Ensure path matches your folder structure

# Fix the Date Format
# CSVs store dates as "text". We must tell Python this column is "Time".
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

print("Setup Complete. Data Ready.")
```

```python
# Create a blank canvas (10 inches wide, 5 inches tall)
plt.figure(figsize=(10, 5))

# Filter for just ONE machine (Plotting 2 machines on 1 line is messy)
press_data = df[df['Machine_ID'] == 'PRESS_01']

# Draw the line
# X-axis = Time, Y-axis = Cycle Time
plt.plot(press_data['Timestamp'], press_data['Cycle_Time_Sec'], marker='o', linestyle='-', color='blue')

# Label the chart (Always label your axes!)
plt.title('Cycle Time Stability - PRESS_01')s
plt.xlabel('Time of Day')
plt.ylabel('Cycle Time (Seconds)')
plt.grid(True) # Add gridlines for readability

# Show the image
plt.show()
```

```python
# Calculate totals
production_counts = df.groupby('Machine_ID')['Parts_Produced'].sum()

# Draw a Bar Chart
production_counts.plot(kind='bar', color=['green', 'orange'])

plt.title('Total Production by Machine')
plt.ylabel('Parts Produced')
plt.xticks(rotation=0) # Keep the text horizontal

plt.show()
```
